```python
import sys
import csv
import time
sys.setrecursionlimit(10**6)  

def dosyadan_veri_oku(filename):
    x_koordinati = []
    y_koordinati = []

    with open(filename) as csvfile:
        reader = csv.reader(csvfile, delimiter=' ')
        next(reader)
        for satir in reader:
            x_koordinati.append(float(satir[0]))
            y_koordinati.append(float(satir[1]))

    return x_koordinati, y_koordinati

def en_yakin_sehir(sehir):
    global sehir_sayisi, sehirler_arasindaki_maliyet, gidilen_sehir, toplam_maliyet
    
    gidilecek_sehir_maliyet = float('inf')
    siradaki_sehir = 0
    
    for i in range(sehir_sayisi):
        if sehirler_arasindaki_maliyet[sehir][i] != 0 and gidilen_sehir[i] == 0:
            if sehirler_arasindaki_maliyet[sehir][i] < gidilecek_sehir_maliyet:
                gidilecek_sehir_maliyet = sehirler_arasindaki_maliyet[sehir][i]
                siradaki_sehir = i
    
    if gidilecek_sehir_maliyet != float('inf'):
        toplam_maliyet = toplam_maliyet + gidilecek_sehir_maliyet
    
    return siradaki_sehir

def sehirler_arasi_uzaklik_hesapla(x_koordinati, y_koordinati):
    sehirler_arasindaki_maliyet = []
    sehir_sayisi = len(x_koordinati)

    for i in range(sehir_sayisi):
        satir = []
        for j in range(sehir_sayisi):
            uzaklik = uzaklik_hesapla(x_koordinati[i], y_koordinati[i], x_koordinati[j], y_koordinati[j])
            satir.append(uzaklik)
        sehirler_arasindaki_maliyet.append(satir)

    return sehirler_arasindaki_maliyet

def uzaklik_hesapla(x1, y1, x2, y2):
    return ((x1 - x2) ** 2 + (y1 - y2) ** 2) ** (1/2)

def en_kisa_yol(sehir):
    global x_koordinati, y_koordinati, gidilen_sehir
    global toplam_maliyet, sehir_sayisi, sehirler_arasindaki_maliyet
    
    gidilen_sehir[sehir] = 1
    gidilecek_sehir = 0
    
    if sehir != 0:
        print(" ", end="")
    
    print(f"{sehir}", end="")
    gidilecek_sehir = en_yakin_sehir(sehir)
    
    if sum(gidilen_sehir) == sehir_sayisi:
        print(" 0")
        toplam_maliyet = toplam_maliyet + sehirler_arasindaki_maliyet[sehir][0]
        return

    en_kisa_yol(gidilecek_sehir)

def main():
    global sehir_sayisi, sehirler_arasindaki_maliyet, gidilen_sehir, toplam_maliyet, x_koordinati, y_koordinati
    
    x_koordinati, y_koordinati = dosyadan_veri_oku("tsp_5_1.csv")
    sehirler_arasindaki_maliyet = sehirler_arasi_uzaklik_hesapla(x_koordinati, y_koordinati)
    
    sehir_sayisi = len(x_koordinati)
    
    gidilen_sehir = []
    for i in range(sehir_sayisi):
        gidilen_sehir.append(0)
    
    toplam_maliyet = 0
    
    #baslangic_zamani=time.time()
    print(f"Optimal maliyeti sağlayan path: ", end="")
    en_kisa_yol(0)
    
    lastValue = toplam_maliyet + sehirler_arasindaki_maliyet[sehir_sayisi-1][0]
    
    print(f"Optimal maliyet değeri: {toplam_maliyet}")
    #bitis_zamani=time.time()
    
    #calisma_zamani = bitis_zamani - baslangic_zamani
    
    #print(f"Çalışma Zamanı: {calisma_zamani:.9f} saniye")
    
if __name__ == '__main__':
    main()
```
